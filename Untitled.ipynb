
Dictionary Attack on Hashed Password
Hashing Algorithm: SHA-256
No database used


```python
import hashlib

# Step 1: Take password input from user
password = input("Enter a password: ")

# Hash the password using SHA-256
hashed_password = hashlib.sha256(password.encode()).hexdigest()

print("\nHashed Password:", hashed_password)

# Step 2: Read dictionary file
found = False

try:
    with open("wordlist.txt", "r") as file:
        for word in file:
            word = word.strip()

            # Step 3: Hash each dictionary word
            hashed_word = hashlib.sha256(word.encode()).hexdigest()

            # Step 4: Compare hashes
            if hashed_word == hashed_password:
                print("\nPassword cracked!")
                print("Password is:", word)
                found = True
                break

    if not found:
        print("\nPassword not found in dictionary.")

except FileNotFoundError:
    print("Wordlist file not found!")
    
```
